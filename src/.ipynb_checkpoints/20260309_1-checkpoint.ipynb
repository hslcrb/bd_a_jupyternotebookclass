## **국영수 과목 점수 평균 구하기 프로그램 코드**

```python
# 정수형(int)으로 변환하여 입력을 받습니다.
kor = int(input("국어 과목의 점수를 입력하십시오: "))
eng = int(input("영어 과목의 점수를 입력하십시오: "))
math = int(input("수학 과목의 점수를 입력하십시오: "))

# 합계와 평균 계산
total = kor + eng + math
average = total / 3

print(f"평균 점수: {average}")
```

## **(확률, 통계값 구하는 머신)** Claude.ai에서 제작함.

```python
print("=== 통계 계산기 | 종료: q ===\n")

while True:
    raw = input("숫자 입력 (공백/쉼표 구분) >> ").strip()
    
    if raw.lower() == 'q':
        print("종료합니다.")
        break
    
    try:
        nums = [float(x) for x in raw.replace(",", " ").split()]
        assert nums, "숫자를 하나 이상 입력하세요."
    except (ValueError, AssertionError) as e:
        print(f"[오류] {e}\n")
        continue

    n    = len(nums)
    mean = sum(nums) / n
    var  = sum((x - mean) ** 2 for x in nums) / n

    # 중앙값
    s = sorted(nums)
    median = (s[n//2 - 1] + s[n//2]) / 2 if n % 2 == 0 else s[n//2]

    # 최빈값: 각 값의 등장 횟수 집계
    freq = {x: nums.count(x) for x in nums}
    max_freq = max(freq.values())
    modes = [k for k, v in freq.items() if v == max_freq]
    mode = modes if len(modes) < n else "없음"  # 모두 빈도 같으면 없음

    print(f"""
  데이터 수 : {n}
  평균      : {mean:.4f}
  분산      : {var:.4f}
  표준편차  : {var**0.5:.4f}
  중앙값    : {median}
  최빈값    : {mode}
  최솟값    : {min(nums)}
  최댓값    : {max(nums)}
""")
```
