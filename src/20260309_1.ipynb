## **(확률, 통계값 구하는 머신)** Claude.ai에서 제작함.

```python
print("=== 통계 계산기 | 종료: q ===\n")

while True:
    raw = input("숫자 입력 (공백/쉼표 구분) >> ").strip()
    
    if raw.lower() == 'q':
        print("종료합니다.")
        break
    
    try:
        nums = [float(x) for x in raw.replace(",", " ").split()]
        assert nums, "숫자를 하나 이상 입력하세요."
    except (ValueError, AssertionError) as e:
        print(f"[오류] {e}\n")
        continue

    n    = len(nums)
    mean = sum(nums) / n
    var  = sum((x - mean) ** 2 for x in nums) / n

    # 중앙값
    s = sorted(nums)
    median = (s[n//2 - 1] + s[n//2]) / 2 if n % 2 == 0 else s[n//2]

    # 최빈값: 각 값의 등장 횟수 집계
    freq = {x: nums.count(x) for x in nums}
    max_freq = max(freq.values())
    modes = [k for k, v in freq.items() if v == max_freq]
    mode = modes if len(modes) < n else "없음"  # 모두 빈도 같으면 없음

    print(f"""
  데이터 수 : {n}
  평균      : {mean:.4f}
  분산      : {var:.4f}
  표준편차  : {var**0.5:.4f}
  중앙값    : {median}
  최빈값    : {mode}
  최솟값    : {min(nums)}
  최댓값    : {max(nums)}
""")
```

## **사용자로부터 각종 조건을 받아 숫자를 어디서부터 어디까지 출력하는 범용 파이썬 코드 from Gemini**

```python
# 사용자로부터 각종 조건 입력받기
start = int(input("시작할 숫자를 입력하세요 (마이너스 가능): "))
end = int(input("어디까지 검사할까요? (끝값): "))
step = int(input("숫자를 몇씩 증가시킬까요? (보통 1): "))
target = int(input("어떤 수의 배수를 찾을까요?: "))

print(f"\n--- {start}부터 {end}까지 {target}의 배수 목록 ---")

# for문과 range를 사용하여 입력받은 범위만큼 반복
# range(시작, 끝+1, 증가폭)을 사용하여 사용자가 원하는 대로 숫자를 키워나갑니다.
for i in range(start, end + 1, step):
    # 입력받은 target의 배수인지 확인 (나머지가 0인 경우)
    if i != 0 and i % target == 0:
        print(i, end=" ")

print("\n-------------------------------------------")
```

## **input()과 for문을 활용해서 원하는 크기의 다이아몬드를 만드는 코드, from Gemini**

```python
height = int(input("다이아몬드의 중간층 길이를 입력하세요: "))

# 1. 윗부분 (삼각형)
for i in range(1, height + 1):
    # 공백 출력: 오른쪽으로 밀어주기
    print(" " * (height - i), end="")
    # 별 출력: 홀수 개수로 늘어남 (1, 3, 5...)
    print("*" * (2 * i - 1))

# 2. 아랫부분 (역삼각형)
for i in range(height - 1, 0, -1):
    # 공백 출력: 다시 줄어들 때 공백은 늘어남
    print(" " * (height - i), end="")
    # 별 출력: 다시 줄어듦
    print("*" * (2 * i - 1))
```

## 이것은 내가 **그냥 하고싶어서** 하는 것(바로 아래의 것) - 날씨 콘솔 프로그램 코드

```python
import urllib.request
import json
import time

def get_weather():
    url = "https://api.open-meteo.com/v1/forecast?latitude=37.26&longitude=127.02&current_weather=true"
    try:
        # 인증서 문제 방지를 위해 간단하게 호출
        with urllib.request.urlopen(url, timeout=5) as response:
            if response.status == 200:
                data = json.loads(response.read().decode('utf-8'))
                current = data.get("current_weather", {})
                return {
                    "temp": current.get("temperature"),
                    "wind": current.get("windspeed"),
                    "time": current.get("time")
                }
    except Exception as e:
        return f"에러 발생: {e}"
    return None

def main():
    print("실시간 날씨 모니터링 시작 (종료: Ctrl+C)")
    
    try:
        while True:
            weather = get_weather()
            
            # ANSI Escape Code: \033[H (커서를 홈으로), \033[J (화면 아래쪽 삭제)
            # 화면 전체를 새로 그리는 게 아니라 커서 위치만 옮겨서 깜빡임을 최소화함
            print("\033[H\033[J", end="") 
            
            print("====================================")
            print("   수원(Suwon) 실시간 날씨 데이터   ")
            print("====================================")
            
            if isinstance(weather, dict):
                print(f" 현재 온도 : {weather['temp']}°C")
                print(f" 현재 풍속 : {weather['wind']} km/h")
                print(f" 업데이트 : {weather['time']}")
            else:
                print(f" 데이터 로드 실패: {weather}")
                
            print("------------------------------------")
            print(" 1초마다 자동 갱신 중... (Ctrl+C 종료)")
            
            time.sleep(1)
            
    except KeyboardInterrupt:
        print("\n\n모니터링을 종료합니다. 열공해라.")

if __name__ == "__main__":
    main()
```

## 1초마다 반복이 아닌, **한번만 나오게 하고 더욱 상세한 정보 출력**

```python
import urllib.request
import json
from datetime import datetime

def get_detailed_weather():
    # 수원 위경도 및 상세 데이터 요청 (습도, 체감온도 등 추가)
    url = "https://api.open-meteo.com/v1/forecast?latitude=37.26&longitude=127.02&current=temperature_2m,relative_humidity_2m,apparent_temperature,is_day,precipitation,weather_code,wind_speed_10m&timezone=Asia%2FSeoul"
    
    try:
        with urllib.request.urlopen(url, timeout=5) as response:
            data = json.loads(response.read().decode('utf-8'))
            current = data.get("current", {})
            units = data.get("current_units", {})
            
            # 날씨 코드를 한글 상태로 변환 (WMO Code 기준)
            wmo_codes = {
                0: "맑음 (Clear sky)",
                1: "대체로 맑음", 2: "구름 조금", 3: "흐림",
                45: "안개", 48: "이슬 안개",
                51: "가랑비", 53: "보통 가랑비", 55: "진한 가랑비",
                61: "약한 비", 63: "보통 비", 65: "강한 비",
                71: "약한 눈", 73: "보통 눈", 75: "강한 눈",
                95: "뇌우"
            }
            status = wmo_codes.get(current.get("weather_code"), "정보 없음")
            
            print("\n" + "="*40)
            print(f" 📍 측정 지점: 경기도 수원시 (한봄고 인근)")
            print(f" ⏰ 확인 시간: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
            print("="*40)
            
            print(f" 🌡️  현재 온도 : {current.get('temperature_2m')}{units.get('temperature_2m')}")
            print(f" 🌡️  체감 온도 : {current.get('apparent_temperature')}{units.get('apparent_temperature')}")
            print(f" 💧 현재 습도 : {current.get('relative_humidity_2m')}{units.get('relative_humidity_2m')}")
            print(f" 🌦️  날씨 상태 : {status}")
            print(f" 🌧️  강수량    : {current.get('precipitation')}{units.get('precipitation')}")
            print(f" 💨 현재 풍속 : {current.get('wind_speed_10m')}{units.get('wind_speed_10m')}")
            print(f" ☀️  낮/밤 구분: {'낮(Day)' if current.get('is_day') == 1 else '밤(Night)'}")
            
            print("="*40)
            print(" [알림] 데이터 출처: Open-Meteo (No API Key)")
            print("="*40 + "\n")

    except Exception as e:
        print(f"\n [!] 데이터를 가져오는 중 에러 발생: {e}")

if __name__ == "__main__":
    get_detailed_weather()
```
