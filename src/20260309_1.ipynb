## 국영수 과목 점수 평균 구하기 프로그램 코드

```python
# 정수형(int)으로 변환하여 입력을 받습니다.
kor = int(input("국어 과목의 점수를 입력하십시오: "))
eng = int(input("영어 과목의 점수를 입력하십시오: "))
math = int(input("수학 과목의 점수를 입력하십시오: "))

# 합계와 평균 계산
total = kor + eng + math
average = total / 3

print(f"평균 점수: {average}")
```
